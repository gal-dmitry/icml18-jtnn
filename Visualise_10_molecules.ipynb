```python
import csv
import matplotlib.pyplot as plt
import os
import numpy as np
from torch.utils.data import Dataset, DataLoader

from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import DrawingOptions

from rdkit.Chem import AllChem
```

**Загружаем сгенерированные SMILES**

```python
with open("molvae/10_molecules.csv", 'r') as File:
    reader = csv.reader(File)
    head = next(reader)
    value = head[0]
    
    smiles = []
    for i, smile in enumerate(reader):
        smiles.append(smile[0])
```

```python
smiles
```

**Строим молекулы по SMILES**

```python
molecules = [Chem.MolFromSmiles(smile) for smile in smiles]
```

```python
molecules
```

**Рисуем картинки по молекулам и сохраняем их в папку images**

```python
for i, mol in enumerate(molecules):
    fig = Draw.MolToMPL(mol)
    fig.gca().set_axis_off()
    fig.savefig('{0}{1}.png'.format('images/', str(i).zfill(2)), bbox_inches='tight')
    plt.close(fig)
```

**Загружаем и отображаем картинки**

```python
class MolDataset(Dataset):
    def __init__(self, path_to_dataset):
        self.names = os.listdir(path_to_dataset)
        self.names.sort()
        self.path = path_to_dataset
    
    def __getitem__(self, index):
        full_path = self.path + "/" + self.names[index]
        img = plt.imread(full_path)
        return np.array(img)

    def __len__(self):
        return len(self.photo_names)
```

```python
def visualize(dataset, indx):
    imgs = [dataset[i] for i in indx] 
    size = len(imgs)
    
    fig = plt.figure(figsize=(10*size, 5*size))
    columns = 1
    rows = size
    ax = []
    
    for num, img in enumerate(imgs):
        ax.append(fig.add_subplot(rows, columns, num+1))
        plt.imshow(img)
        plt.axis('off') 

    plt.tight_layout()
```

```python
visualize(MolDataset('images'), range(10))
```
